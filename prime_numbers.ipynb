```python
import matplotlib.pyplot as plt
import math as m
```

```python
def sum_dividers(number: int) -> int: 
    dividers_sum = 1
    for div in range(2,number):
        if number % div == 0:
            dividers_sum += div
            
    return dividers_sum
```

```python
n = 10

x = list(range(0, n, 1))
y = [sum_dividers(item) for item in x]
```

```python
def is_prime(x: list, y: list):
    x_prime = []
    y_prime = []
    for i in range(0, len(x)):
        if x[i] == y[i]:
            x_prime.append(x[i])
            y_prime.append(y[i])
    
    return x_prime, y_prime
```

```python
def is_simple(x: list, y: list):
    x_simple = []
    y_simple = []
    for i in range(0, len(x)):
        if y[i] == 1:
            x_simple.append(x[i])
            y_simple.append(y[i])
    
    return x_simple, y_simple
```

```python
x_prime, y_prime = is_prime(x, y)
y_prime
```

```python
x_simple, y_simple = is_simple(x, y)
x_simple
```
