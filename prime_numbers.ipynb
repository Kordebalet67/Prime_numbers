```python
import matplotlib.pyplot as plt
import math as m
```

```python
def sum_dividers(number: int) -> int: 
    dividers_sum = 1
    for div in range(2,number):
        if number % div == 0:
            dividers_sum += div
            
    return dividers_sum
```

```python
n = 1_000

x = list(range(0, n, 1))
y = [sum_dividers(item) for item in x]
```

```python
def is_prime(x: list, y: list):
    x_prime = []
    y_prime = []
    for i in range(0, len(x)):
        if x[i] == y[i]:
            x_prime.append(x[i])
            y_prime.append(y[i])
    
    return x_prime, y_prime
```

```python
def is_simple(x: list, y: list):
    x_simple = []
    y_simple = []
    for i in range(0, len(x)):
        if y[i] == 1:
            x_simple.append(x[i])
            y_simple.append(y[i])
    
    return x_simple, y_simple
```

```python
x_prime, y_prime = is_prime(x, y)
y_prime
```

```python
x_simple, y_simple = is_simple(x, y)
x_simple
```

```python
import plotly.express as px
import pandas as pd

prime = ["is prime" for i in range (0, len(x_prime))] 
data_prime = {'x_prime': x_prime, 'y_prime': y_prime, 'is_prime': prime}
df_prime = pd.DataFrame(data = data_prime)

fig = px.scatter(df_prime, x="x_prime", y="y_prime", hover_data=["is_prime"])

simple = ["is prime" for i in range (0, len(x_simple))]
data_simple = {'x_simple': x_simple, 'y_simple': y_simple, 'is_simple': simple}
df_simple = pd.DataFrame(data = data_simple)

fig = px.scatter(df_simple, x="x_simple", y="y_simple", hover_data=["is_simple"])
fig = plt.figure(figsize=(36, 12))


#fig_prime.patch.set_facecolor('grey')
ax = fig.add_subplot()
ax.patch.set_facecolor('grey')
ax.patch.set_alpha(0.5)



plt.plot(x, y, 'ok')
plt.plot(x_prime, y_prime, 'or')
plt.plot(x_simple, y_simple, 'og')
# включаем основную сетку
plt.grid(which='major')
# включаем дополнительную сетку
plt.grid(which='minor', linestyle=':')

plt.semilogy()

plt.show()
```
